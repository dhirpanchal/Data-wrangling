```python
# Importing important libraries
import seaborn as sns
```

```python
# catplot
exercise = sns.load_dataset('exercise')
a = sns.catplot(x = 'time', y = 'pulse', hue = 'kind', data = exercise)
```

```python
# hue : encodes the points with different colors with respect to target variable.
```

```python
# distplot
df = sns.load_dataset('tips')
df.head()
```

```python
# distplot : used basically for univariate set of observations and visualizing them thrrough histogram.
```

```python
# distplot
sns.distplot(df['total_bill'], kde = False, color = 'red', bins = 10)
```

```python
# distplot
sns.distplot(df['total_bill'], kde = True, color = 'green', bins = 10)
```

```python
# Pair Plot
sns.pairplot(df, hue = 'sex', palette = 'coolwarm')
```

```python
# Pair Plot
sns.pairplot(df, hue = 'smoker', palette = 'coolwarm')
```

```python
# jointplot
sns.jointplot(x = 'total_bill', y = 'tip', data = df)
```

```python
# jointplot
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'kde')
```

```python
# jointplot
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex')
```

```python
# jointplot
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
# kind can have hex, reg, kde values.
# Default for kind is scatter.
```
